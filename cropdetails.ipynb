```python
#Reading csv file and converting it to point shapefile form latitude and longitude fields
import pandas as pd
hos=gp.read_file(r"C:\Users\Ahammad Sahil\Desktop\SEM 2\GA LAB\anju\pylab1\Sample_Data_Geopandas_2022\Hospitals.csv")
hos
```
